```python
#1- Qual a quantidade de incidentes por gênero?
qtd_gen = df.groupby('victims_sex').agg({'incident_event_type':'count'}).sort_values(by= 'incident_event_type', ascending= False).reset_index()
```

```python
sns.set_style("whitegrid")
sns.catplot(
    data=qtd_gen, kind="bar",
    x="victims_sex", y="incident_event_type",
    errorbar="sd", palette="dark", alpha=.6, height=6
).set_axis_labels("Gênero","Número de incidentes").despine(left=True,bottom= True)
plt.title("Quantidade de Incidentes por Gênero")
```

### Percebe-se uma quantidade de vítimas em acidentes relacionando o gênero feminino bem maior em relação ao gênero masculino, dentro dos incidentes onde foram identificados os gêneros das vítimas.

```python
#2- Qual faixa etária esteve mais envolvida nos incidentes?

faixa_etaria = df.groupby('victims_age').agg({'victim_category':'count'}).reset_index()
faixa_etaria
```

```python
sns.set_style("whitegrid")
sns.catplot(
    data=faixa_etaria, kind="bar",
    x="victims_age", y="victim_category",
    errorbar="sd", palette="pastel", alpha=.6, height=6
).set_axis_labels("Faixa etária","Número de incidentes").despine(left=True,bottom= True)
plt.title("Quantidade de Incidentes por Faixa Etária")
```

### A faixa etária que se apresenta como a que mais se envolve em acidentes é a adulta, que pode ser justificada pela grande quantidade de adultos ter uma necessidade maior de participação em atividades de risco, como trabalho em construção civil ou direção de veículos, comportamentos de risco no trânsito, como excesso de velocidade ou uso de álcool ao dirigir,lazer ou resolver problemas no dia a dia.
### Crianças, adolecentes e idosos podem ser envolvidas em uma menor quantidade de acidentes devido a menor tempo de exposição a situações de risco, como no caso de crianças, menor capacidade física e reflexos mais lentos, no caso de idosos, o que pode levar a uma maior cautela em atividades de risco.

### Os incidentes menos comuns são provenientes de Falha cítica de segurança, Incidente de atividade, Vandalismo e Hooliganismo, Roubo, Incendio e Assaltos.

### Escorregão, tropeção e queda podem ser frequentes devido a:
### - Falta de atenção ao ambiente e condições precárias do local ou de circulação.
### - Calçados inadequados ou piso escorregadio.
### Lesões a bordo, Lesões pessoais e Colisão podem ser frequente em ambientes com grande fluxo de pessoas ou veículos, como:
### - Lojas, fábricas, ruas movimentadas.
### - Falta de atenção ou imprudência dos envolvidos.


```python
#4- Como foi a evolução de incidentes por mês ao longo do tempo?
n_df = df
```

```python
df['date_of_incident'] = pd.to_datetime(df['date_of_incident'], format='%Y-%m-%d')
```

```python
df['date_of_incident'] = df['date_of_incident'].dt.strftime('%Y/%m/%d')
```

```python
incidentes_tempo = df.groupby('date_of_incident').agg({'incident_event_type':'count'}).reset_index()
incidentes_tempo
```

```python
sns.set(style="whitegrid")

# Criando o gráfico de linha
plt.figure(figsize=(10, 6))
sns.lineplot(data=incidentes_tempo, x='date_of_incident', y='incident_event_type', linewidth = 1)
plt.plot([incidentes_tempo['date_of_incident'].min(), incidentes_tempo['date_of_incident'].max()], 
         [incidentes_tempo['incident_event_type'].min(), incidentes_tempo['incident_event_type'].max()],
         color='red', linestyle='--')
plt.title('Evolução de Incidentes por Mês ao Longo do Tempo')
plt.xlim()
plt.xlabel('Data')
plt.ylabel('Número de Incidentes')
plt.xticks(rotation = 45)
plt.xticks(incidentes_tempo['date_of_incident'].tolist()[0::6])
plt.show()
```

### O gráfico apresenta uma tendência de aumento dos incidentes ao longo do tempo, tendo algumas quedas pontuais.
### Alguns picos de incidentes são observados em: 
### Junho de 2015: com cerca de 560 incidentes;
### Novembro de 2016: com cerca de 610 incidentes;
### Julho de 2017: com cerca de 650 incidentes;
### Maio de 2018: com cerca de 580 incidentes. 
### Os menores números de incidentes são observados em:
### Fevereiro de 2015: com cerca de 360 incidentes;
### Março de 2016: com cerca de 450 incidentes;
### Janeiro de 2017: com cerca de 440 incidentes;
### Agosto de 2018: com cerca de 490 incidentes.
### Possíveis Explicações:

### Fatores Sazonais: É possível que alguns meses do ano apresentem um aumento natural no número de incidentes, devido a fatores como:
### Festas e eventos com grande aglomeração de pessoas.
### Condições climáticas adversas.
### Eventos Específicos: Eventos específicos podem ter gerado um aumento no número de incidentes em determinados meses, como:
### Falhas pontuais em medidas de segurança ou falhas humanas.
### Desastres naturais ou outros eventos de grande impacto.

```python
#8- No ano de 2017 em qual mês houve mais incidentes com idosos (Elderly)?
df
```

```python
locale.setlocale(locale.LC_TIME, '')
```

```python
elderly_2017 = df[(df['victims_age'] == 'Elderly') & (df['year'] == 2017)]
```

```python
pd.to_datetime(elderly_2017['date_of_incident']).dt.strftime('%B')
```

```python
elderly_2017['month'] = pd.to_datetime(elderly_2017['date_of_incident']).dt.strftime('%B')
```

```python
meses_ordenados = [
    'janeiro', 'fevereiro', 'março', 'abril', 'maio', 'junho',
    'julho', 'agosto', 'setembro', 'outubro', 'novembro', 'dezembro'
]
```

```python
elderly_2017
```

```python
elderly_2017['ordem_mes'] = pd.Categorical(elderly_2017['month'], categories=meses_ordenados, ordered=True)
```

```python
elderly_2017_gr = elderly_2017.groupby('ordem_mes').agg({'victims_sex':'count'}).reset_index().sort_values(by = 'ordem_mes', ascending= True)
```

```python
elderly_2017_gr['victims_sex'].mean()
```

```python
elderly_2017_gr
```

```python
sns.set_style("whitegrid")
sns.catplot(
    data=elderly_2017_gr, kind="bar",
    x="victims_sex", y="ordem_mes",
    errorbar="sd", palette="flare", alpha=.6, height=6
).set_axis_labels("Quantidade de Incidentes","Data do incidente").despine(left=True,bottom= True)
plt.title("Quantidade de Incidentes com Idosos no ano de 2017")
```

### Ao analisar o gráfico, percebe-se que no ano de 2017, o mês onde ocorreu mais incidentes foi em Julho, com uma quantidade de 81 acidentes e tendo uma média mensal de 70 acidentes.
